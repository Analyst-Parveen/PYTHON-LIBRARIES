```python
#  BAR-PLOT:=>
# A bar plot shows how different categories compare using bars to represent their values. 

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
print(data)
sequence=['Sun','Thur','Fri','Sat']
sns.barplot(data=data,x='day',y='tip',estimator='sum',hue='sex',palette='Blues',errorbar=('ci', 50),order=sequence)
plt.show()
```

```python
# HISTOGRAM:=>TO COUNT THE FREQUENCY OF THE DATA IN A FIX RANGE.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.histplot(data=data,x='day',hue='sex',kde='true')
plt.show()
                      
                      
```

```python
# SCATTER PLOT:=> USE WHERE FINF NO TO NO RELATIONSHIP OR NUERIC TO NUMERIC.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)

sns.scatterplot(data,x='total_bill',y='tip',hue='smoker',marker='s',size='size',palette='spring')
plt.legend(bbox_to_anchor=(0.2,1,0,1))
plt.show()
```

```python
#  HEAT-MAP:=> ALSO USE AS SCATTE RPLOT AND ITS ALSO USE WHERE THE BOTH AXIS ARE NUMERIC VALUE.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
gp=data.groupby('day').agg({'tip':'mean'})
print(gp)
sns.heatmap(gp)
plt.show()
```

```python
# COUNT-PLOT;=> TO USE THE COUNT OF A VALUE WE USE COUNT PLOT.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.countplot(data,x='sex',hue='day',palette='GnBu')
plt.show()
```

```python
#  VIOLIN_PLOT:=> TO CEHCK THE FREQUENCY OF DATA.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.violinplot(data,x='total_bill')
```

```python
# STRIP-PLOT:=> SAME AS SCATTER PLOT BUT WE CAN USE STRING VALUE AT ONE AXIS.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.stripplot(data,x='day',y='total_bill',hue='sex',dodge=True,jitter=0.2)
plt.show()
```

```python
#  BOX-PLOT:=> TO FIND THE STATICAL ANALYSIS.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
# sns.boxplot(data,x='tip',fliersize=10)
sns.boxplot(data,x='day',y='tip',fliersize=10,orient='v')
plt.show()
```

```python
# CAT-PLOT:=> ITS DEAFULT VALUE IS STRIP PLOT. ITS SIMILAR TO THE STRIP PLOT but can create diff diff chart bar,violoin etc.


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.catplot(data,x='day',y='tip',hue='sex',kind='bar')
#  for count plot remove y and kind=count.--------------------

plt.show()

```

```python
#  PAIR-PLOT:=> PAIR PLOT USING A MULTIPLE DATA BETWEEN NUMERIC COLUMN.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.pairplot(data,diag_kind='hist',kind='kde',hue='day')
plt.show()
```

```python
#  STYLE AND COLOR IN SEABORN:=>

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('exercise')
# print(data)
sns.barplot(data,x='time',y='pulse')
sns.set_style(style='darkgrid')
sns.palplot(sns.color_palette('spring',3))

plt.show()
```

```python
#  SWARM PLOT:=> IT IS SIMILAR TO THE STRIP PLOT BUT THER IS N OVERLAPPING OF DOTS IN GRAPHS BUT IN STRIP CAN.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.swarmplot(data,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
#  KDE PLOT:=> KERNAL DENSITY ESTIMATION.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.kdeplot(data,x='total_bill',hue='day',multiple='stack')
plt.show()
```

```python
#  RELATION-PLOT:=> it is use to find teh relation b/w two data points.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
sns.relplot(data,x='tip',y='total_bill',hue='sex',kind='scatter',col='day',size='smoker')
```

```python
#  MULTIPLE-PLOT:=> TO USE CRATE MULTIPLE PLOT IN A LINE.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=sns.load_dataset('tips')
# print(data)
a=sns.FacetGrid(data,col='day',height=2)
a.map(sns.barplot,'tip','smoker')
plt.show()
```
